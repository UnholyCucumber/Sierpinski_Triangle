# Sierpiński Triangle

Named after the Polish mathematician Wacław Sierpiński; this is one of the simplest, yet most elegant self-similar sets. As an avid fan of the *Legend of Zelda* video game franchise growing up, this pattern has special significance for me :)

### There are three transformations:

First Transformation:

$$x_{n+1} = 0.5x_n$$
$$y_{n+1} = 0.5y_n$$

Second Transformation:

$$x_{n+1} = 0.5x_n + 0.5$$
$$y_{n+1} = 0.5y_n + 0.5$$

Third Transformation:
$$x_{n+1} = 0.5x_n + 1$$
$$y_{n+1} = 0.5y_n$$

```python
import matplotlib.pyplot as plt
from random import choice

def first_transformation(p):
    x0 = p[0]
    y0 = p[1]
    x1 = 0.5 * x0
    y1 = 0.5 * y0
    return x1, y1

def second_transformation(p):
    x0 = p[0]
    y0 = p[1]
    x1 = 0.5 * x0 + 0.5
    y1 = 0.5 * y0 + 0.5
    return x1, y1

def third_transformation(p):
    x0 = p[0]
    y0 = p[1]
    x1 = 0.5 * x0 + 1
    y1 = 0.5 * y0
    return x1, y1

transformations = [first_transformation, second_transformation, third_transformation]
```

Let's try running it with just 1 iteration.

```python
# initialize the pattern generation
x_coords = [0]
y_coords = [0]
x, y = 0, 0

for i in range(1):
    transform = choice(transformations)
    x, y = transform((x, y))
    x_coords.append(x)
    y_coords.append(y)

plt.rc('figure', figsize = (8, 8))
plt.plot(x_coords, y_coords, 'o')
```
